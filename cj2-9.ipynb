```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# pull uniform distribution of 1000 random numbers between 0 and 1
x = np.random.random(1000)
print(x)
```

```python
# creating a figure and axes

f, ax = plt.subplots(1,1,figsize=(10,5))

ax.hist(x,bins=100,edgecolor='black',alpha=0.5)

ax.set_xlabel('X')
ax.set_ylabel('N per Bin')

plt.savefig('cj2-9_plot.pdf')
```
