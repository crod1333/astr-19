```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def exp(x):
    # takes e (euler's number: ~2.7) and raises it to the x power
    return np.exp(x)
```

```python
# creating data x (100 floating numbers from 0 to 1, inclusive)

x = np.linspace(0,1,100)
print(x)

# making the function exp(x) into a variable

y = exp(x)
print(y)
```

```python
f, ax = plt.subplots(1,1,figsize=(5,5))

# plotting x vs y
ax.plot(x,y,linewidth=2)
ax.set_xlabel('Time [milliseconds]')
ax.set_ylabel('Awesomeness')

plt.savefig('cj2-7_plot.pdf')
```
