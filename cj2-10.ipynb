```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# np.random.randint
    # can set a low and high value for the distribution that you want to pull random numbers from
    # can set a size for the number of values you want; you could also provide a tuple

x = np.random.randint(low=0,high=101,size=1000)
```

```python
# create a figure (canvas)
f = plt.figure(figsize=(5,5))

#creating 100 bins for the data
plt.hist(x, bins=100, edgecolor='black',alpha=0.5)

plt.xlabel('X')
plt.ylabel('N per Bin')
plt.title('Random numbers distributed into bins')

plt.savefig('cj2-10_plot.pdf')
```
