```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# data to implement into functions

x = np.linspace(0,1,100)
```

```python
# define sin and cos functions

def sin(x):
    return np.sin(x)

def cos(x):
    return np.cos(x)

y = sin(x)
z = cos(x)
```

```python
f, axarr = plt.subplots(1,2, figsize=(6,5))

# left plot
axarr[0].plot(x,y)
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')
axarr[0].set_title(r'$\sin(x)$')

# right plot
axarr[1].plot(x,z)
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title(r'$\cos(x)$')

# figure method ; adjust wdith size between axes
f.subplots_adjust(wspace=0.4)

plt.savefig
```
